# データ可視化

## データ取得

```python
from database_sql import Database
import datetime
```

```python
db = Database()
```

```python
db.sql = "select per from stock.price where date={} and per>0 order by per asc;".format("'2017/12/1'")
```

```python
per = db.execute()
```

```python
print(per)
```

```python
per_list = []
for p in per:
    per_list.append(p["per"])
```

```python
print(per_list)
```

## プロット

```python
import matplotlib.pyplot as plt
```

```python
x = list(range(0,100))
```

```python
len(list(filter(lambda x: x<3, per_list)))
```

```python
distribution = list(map(lambda th: len(list(filter(lambda per: per < th, per_list))), x))
```

```python
distribution
```

```python
plt.scatter(x, distribution)
plt.show()
```

プロットのレンジを変える

```python
i = [2,20]
```

```python
x = list(range(i[0],i[1]))
distribution = list(map(lambda th: len(list(filter(lambda per: per < th, per_list))), x))
```

```python
plt.scatter(x, distribution)
plt.show()
```

分布を書いてみる

```python
i = [0,100]
```

```python
x = list(range(i[0],i[1]))
distribution = list(map(lambda th: len(list(filter(lambda per:  th-1 <= per < th, per_list))), x))
```

```python
plt.scatter(x, distribution)
plt.show()
```
