# Jupyter Notebook紹介

元はipython notbook、その後ＣやＣ＃などほかの言語にも対応したのでJupyter Notebookとなった。

オープンソース

## インストール

http://jupyter.org/install.html

- 方法1 pythonをインストールしてから、コマンドプロンプトで「pip install jupyter」

- 方法2（公式ページおすすめ） Anacondaはpythonも一緒にインストールしてくれる

## 起動

コマンドプロンプトで「jupyter notebook」

## 使い方

http://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/Notebook%20Basics.html

- 編集モード（edit mode）でセルの編集。

- 操作モード(command mode)でセルの操作や実行

セルには主にコード(code)とマークダウン(markdown)の二種類があり、編集バーで切り替えられる。
（操作モードで「m」を押すとマークダウンモード、「y」を押すとコードモードに切り替え）

その他、編集バーでセルの追加、切り取り、移動、実行などもできる。

**よく使うショートカット(操作モード):**

- 「a」(現在にいるセルの前)と「b」(現在にいるセルの後ろ)にセル追加。
- 「x」でセル削除
-  「Shift+Enter」か「Ctrl+Enter」でセル実行
- 「m」と「y」でセルのcodeとmarkdown切り替え
- 「Enter」でセル編集
- 「Shift+方向キー」で複数セル選択
- （一番重要）「h」：ショートカットキー一覧

**よく使うショートカット(編集モード):**

- 「tab」：自動補完
- ほかいろいろ（VS CodeStudioと似ている感じがする）

**拡張機能（extension）:**

https://github.com/ipython-contrib/jupyter_contrib_nbextensions

- ステップ1, pip install jupyter_contrib_nbextensions
- ステップ2, jupyter contrib nbextension install --user

でインストールすれば、ファイル選択画面へ行くとファイルタブの右端にExtensionタブができる。

おすすめ：

- Collapsible Headings: markdownの見出しごとに畳み込みできる（構造が分かりやすくなる）
- Hinterland：常に自動補完（好き嫌いがあるかも）

# 例：Seleniumを使ってみる

### Seleniumとは

プログラムでブラウザを操作するツール。
主には

- 手軽に使えるSelenium IDE
- Javaやpython、Rucbyなどでコードを書いて動かすSelenium Webdriver

があり、自動テストやスクレイピングの強力なツール。

動的に生成されるページだと、コードが動かなかったり、要素が取りにくかったり、try＆errorで検証しながらコードを書くことも。

### 動かす

ブラウザを立ち上げる

```python
from selenium import webdriver

br = webdriver.Chrome()
```

ウィンドウ最大化

```python
br.maximize_window()
```

URLを開く

```python
url = "https://www.techfirm.co.jp/"
br.get(url)
```

要素を探す

```python
el= br.find_element_by_xpath("//*[contains(text(),'人工知能')]")
```

```python
el.text
```

```python
from selenium.webdriver.common.action_chains import ActionChains
hover = ActionChains(br).move_to_element(el)
hover.perform()
```

```python
br.save_screenshot("img/screenshot.jpg")
```

```python
el.click()
```

Windowを閉じる

```python
br.quit()
```
