# インストール

On Windows, there are no other modules to install.

On OS X, run sudo pip3 install pyobjc-framework-Quartz, sudo pip3 install pyobjc-core, and then sudo pip3 install pyobjc.

On Linux, run sudo pip3 install python3-xlib and sudo apt-get scrot. (Scrot is a screenshot program that PyAutoGUI uses.)

After these dependencies are installed, run pip install pyautogui (or pip3 on OS X and Linux) to install PyAutoGUI.

```python
def save_positions():
    import pyautogui
    pos = []
    desc = "def"
    while desc:
        desc = input()
        pos.append({desc: pyautogui.position()})
    for p in pos:
        print(p)
    return pos
```

# シナリオ作成

```python
def make_scenario():
    import pyautogui
    scenario = []
    while True:
        print("操作：")
        desc = input()
        if desc in ["click"]:
            print("マウスをクリックしたい場所に置いて、リターンを押してください：")
            p = input()
            scenario.append({"click": pyautogui.position()})
        if desc in ["screenshot"]:
            print("マウスを左上の座標に合わせて、リターンを押してください：")
            p = input()
            pos_1 = pyautogui.position()
            print("マウスを右下の座標に合わせて、リターンを押してください：")
            p = input()
            pos_2 = pyautogui.position()
            scenario.append({"screenshot": (pos_1[0],pos_1[1],pos_2[0]-pos_1[0],pos_2[1]-pos_1[1])})
        if desc in ["q",""]:
            for op in scenario:
                print(op)
            print(scenario)
            return scenario
            
```

```python
scenario = make_scenario()
```

# シナリオ実行

```python
def operate_scenario(scenario, file="", wait=1):
    if file:
        pass
    else:
        file="00.jpg"
    for op in scenario:
        key = list(op.keys())[0]
        if key in ["click"]:
            pyautogui.click(op[key])
        if key in ["screenshot"]:
            pyautogui.screenshot(file, region=op[key])
        time.sleep(wait)
```

```python
operate_scenario(scenario)
```
