## 密度分布

```python
def get_density(arr, step):
    i = 0
    density = []
    for item in arr:
        while len(density) <= i:
            density.append(0)
        if item < i*step:
            density[i] += 1
        else:
            density[i] = density[i]/step
            i += 1
    return density
```

```python
def plot_density(arr, step):
    density = get_density(arr, step)
    x = [i*step for i in range(0, len(density))]
    plt.plot(x, density)
    plt.show()
```

```python
get_density(per_list, 0.5)
```

```python
plot_density(per_list, 1)
```
